### 2. Import the Necessary Dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

### 3. Load the Stackoverflow Survey Dataset:

* modify the dataset_path variable if needed
* modify row_count to display the desired amount of rows

```python
dataset_path = "survey_results_public.csv"
data = pd.read_csv(dataset_path)

row_count = 100
data.head(row_count)
```

### 4. Extract Desired Data and Filter by Age Group

* selected columns include: 'ResponseId', 'Age', 'LearnCode', 'YearsCode', 'Country', 'LanguageHaveWorkedWith'
* selected rows include the 'Age' group that is: '18-24 years old'

```python
desired_data = data[['ResponseId', 'Age', 'LearnCode', 'YearsCode', 'Country', 'LanguageHaveWorkedWith']]
desired_data = desired_data[(data['Age'] == '18-24 years old')]

desired_data.head(row_count)
```

### 5. Set 'ResponseId' Column as Index

Since 'ResponseId' is the unique identifier it's logical to add index to this column, in this case ind is used for testing and can be mod ified to test the desired row.

```python
data.set_index('ResponseId', inplace=True)

ind = 15
data.loc[ind]
```

### 6. Convert YearsCode' and 'YearsCodePro' Columns to Numeric

This conversion is necessary to make later comparisons and chart building easier, dataframe column types are printed to make sure, they were indeed converted to floats.

```python
data['YearsCode'] = pd.to_numeric(data['YearsCode'], errors='coerce')
data['YearsCodePro'] = pd.to_numeric(data['YearsCodePro'], errors='coerce')

data.dtypes[['YearsCode', 'YearsCodePro']] 
```

### 7. Filter Data based on Programming Language and Years of Coding Experience

Filtering is done based on:
* if programming languages include Python 
* if coding experience is more than 10 years

```python
new_data = data[(data['LanguageHaveWorkedWith'].str.contains('Python', na=True)) & (data['YearsCode'] >= 10)]
new_data.head(row_count)
```

### 8. Sort Data Based on 'YearsCode' and 'YearsCodePro' in Ascending order

```python
sorted_data = data.sort_values(by=['YearsCode', 'YearsCodePro'], ascending=[True, True])  
sorted_data.head(row_count)
```

### 9. Calculate Statistics for 'YearsCode' Column

Calculating mean, standard, median, min and max using numpy and printing the results rounded to 3 decimals

```python
mean_value = np.mean(data['YearsCode'])
std_deviation = np.std(data['YearsCode'])
median_value = np.median(data['YearsCode'])
min_value = np.min(data['YearsCode'])
max_value = np.max(data['YearsCode'])

print("The following information is calculated based on 'YearsCode' column:")
print("Mean:", mean_value.round(decimals=3))
print("Standard Deviation:", std_deviation.round(decimals=3))
print("Median:", median_value)
print("Min:", min_value)
print("Max:", max_value)
```

### 10. Visualize Distribution of 'Age' using a Bar Chart

Simple bar chart visualising which age groups were most common in this survey.

```python
data['Age'].value_counts().plot(kind='bar', 
                                color="purple")
plt.title('Age Bar Chart:')
plt.xlabel('Age Groups')
plt.xticks(rotation=45)
plt.show()
```

### 11. Visualize Top 10 Most Common Countries using a Bar Chart

Simple bar chart visualising which countries were most common in this survey:
* top 10 countries are retrieved using numpy
* layout is adjusted to prevent overlapping

```python
country_counts = data['Country'].value_counts()

top_countries_indices = np.argsort(country_counts)[::-1][:10]
top_countries = country_counts.index[top_countries_indices]
top_country_counts = country_counts.values[top_countries_indices]

plt.figure(figsize=(10, 6))
plt.bar(np.arange(len(top_countries)), 
        top_country_counts, 
        color='purple')

plt.title('Top 10 Most Common Countries:')
plt.xticks(np.arange(len(top_countries)), 
           top_countries, 
           rotation=45, 
           ha='right') 

plt.tight_layout() 
plt.show()
```

### 12. Create a Heatmap of Converted Compensation for Top 5 Countries and Top 5 Age Groups

* extract top 5 countries and top 5 age groups based on their frequency in the dataset
* heatmap will be focusing on the most common groups for visualization
* display the heatmap based on the converted yearly compensation mean

```python
top_countries = data['Country'].value_counts().head(5).index.tolist()
top_age_groups = data['Age'].value_counts().head(5).index.tolist()

# create chart_data df with top 5 countries and age groups 
chart_data = data[data['Country'].isin(top_countries) & data['Age'].isin(top_age_groups)]

heatmap_data = chart_data.pivot_table(index='Country', 
                                      columns='Age', 
                                      values='ConvertedCompYearly', 
                                      aggfunc='mean')

plt.figure(figsize=(12, 6))  

sns.heatmap(heatmap_data,
            cmap=sns.color_palette("rocket"), 
            annot=True, 
            fmt=".0f", 
            linewidths=.5)  

plt.title('Heatmap of Converted Compensation for Top 5 Countries and Top 5 Age Groups')
plt.show()
```
